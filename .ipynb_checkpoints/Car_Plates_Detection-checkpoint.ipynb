### Import Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

### Open Image

```python
car_img = cv2.imread('Images/car_plate.jpg')
```

### Convert Image

```python
def display(img):
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111)
    new_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    ax.imshow(new_img)
```

### Display it

```python
display(car_img)
```

### Create Classifier

```python
plate_classifier = cv2.CascadeClassifier('Haarcascades/haarcascade_plate_number.xml')
```

### Algo

```python
def detect_license_plates(img):
    plate_image = img.copy()
    license_plates = plate_classifier.detectMultiScale(plate_image, 1.1, 3)
    
    for (x,y,w,h) in license_plates:
        cv2.rectangle(plate_image,
                     (x,y),
                     (x+w, y+h),
                     (255,255,255),
                     3)
    return plate_image 
    
```

### Check the Results

```python
result = detect_license_plates(car_img)
```

```python
display(result)
```
