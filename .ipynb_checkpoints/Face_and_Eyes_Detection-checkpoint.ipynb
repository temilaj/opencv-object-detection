### Import Libraries

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

#render matplotlib images in the same window
%matplotlib inline
```

### Create Classifiers

```python
eye_classifer = cv2.CascadeClassifier('Haarcascades/haarcascade_eye.xml')
face_classifer = cv2.CascadeClassifier('Haarcascades/haarcascade_frontalface_default.xml')
```

### Open Image

```python
image = cv2.imread('Images/eye_face.jpg')
plt.imshow(image)
```

### Fix Image

```python
fixed_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(fixed_image)
```

### Face Classifier

```python
faces = face_classifer.detectMultiScale(fixed_image, 1.3, 5)
```

### If No Face Detected

```python
if faces is ():
    print('No faces found')
```

```python
def detect_eyes_and_faces(img):
    faces_rect = face_classifer.detectMultiScale(img)
        
    for (x, y, w, h) in faces_rect:
        cv2.rectangle(img,
                     (x,y),
                     (x+w, y+h),
                     (255, 0, 0),
                     7)
        
        eyes_rect = eye_classifer.detectMultiScale(img)
        for (ix, iy, iw, ih) in eyes_rect:
            cv2.rectangle(img,
                         (ix,iy),
                         (ix+iw, iy+ih),
                         (0, 0, 255),
                         5)
    
    return img
```

```python
result = detect_eyes_and_faces(fixed_image)
plt.imshow(result)
```
