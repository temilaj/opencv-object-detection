### Import Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#render matplotlib images in the same window
%matplotlib inline
```

### Open Image

```python
image = cv2.imread('Images/eye_face.jpg')
```

### Show Image

```python
plt.imshow(image)
```

### Fix Image

```python
fix_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

### Show Image

```python
plt.imshow(fix_image)
```

### Create Classifier

```python
eye_classifer = cv2.CascadeClassifier('Haarcascades/haarcascade_eye.xml')
```

### It's a Kind of Magic

```python
def detect_eye(img):
    eyes_rects = eye_classifer.detectMultiScale(img)
    
    for (x,y,w,h) in eyes_rects:
        cv2.rectangle(img, 
                     (x,y),
                     (x+w, y+h),
                     (255, 255,255),
                     10)
    return img
```

### Show Results

```python
result = detect_eye(fix_image)
plt.imshow(result)
```
